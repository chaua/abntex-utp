### Estimativa Evolutiva da Raiz Cúbica de um Número Real

image.png

O objetivo deste exercício é utilizar um Algoritmo Genético (AG) para estimar a **raiz cúbica** de um número real positivo $N$. Isto é, encontrar um valor $x \in \mathbb{R}^+$ tal que:

$$
x^3 \approx N
$$

Esse problema, apesar de simples do ponto de vista numérico, será tratado como uma tarefa de otimização contínua, ideal para a aplicação de AGs com representação real.

### Instância de Teste

Utilize $N = 125$. A raiz cúbica exata é $x = 5$.

Sugestões de fitness:
   - ${fitness}(x) = -|x^3 - N|$
   - ${fitness}(x) = - (x^3 - N)^2$
   - ${fitness}(x) = \frac{1}{1 + |x^3 - N|}$
